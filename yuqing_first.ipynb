```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
mydata=pd.read_csv('avgHgt.csv')
df=pd.DataFrame(mydata)
df.columns = ['age','CHeight','JHeight']
df.to_csv('avgHgt.txt',index=0)
plt.plot(df['age'],df['CHeight'],df['age'],df['JHeight'])
plt.title('Height figure')
plt.legend(['CHeight','JHeight'])
plt.xlabel('age')
plt.ylabel('height')
```
